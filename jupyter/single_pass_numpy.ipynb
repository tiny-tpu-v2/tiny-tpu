```python
import numpy as np

# Set NumPy to display all numbers with 4 decimal places
np.set_printoptions(precision=4, suppress=True)

# Learning Rate
lr = 0.75

# Inputs and Outputs of XOR
X = np.array([[0., 0.],
              [0., 1.],
              [1., 0.],
              [1., 1.]])
print(f'X:\n{X}\n')

Y = np.array([[0.],
              [1.],
              [1.],
              [0.]])
print(f'Y:\n{Y}\n')

# Weights and Biases
W1 = np.array([
    [0.2985, -0.5792], 
    [0.0913, 0.4234]
])
print(f'W1:\n{W1}\n')

W2 = np.array([
    [0.5266, 0.2958]
])
print(f'W2:\n{W2}\n')

B1 = np.array([-0.4939, 0.189])
print(f'B1:\n{B1}\n')

B2 = np.array([0.6358])
print(f'B2:\n{B2}\n')
```

## Forward Pass

```python
# Layer 1
Z1 = X @ W1.T
print(f'Z1 pre bias:\n{Z1}\n')
Z1 = Z1 + B1
print(f'Z1:\n{Z1}\n')
H1 = np.maximum(0.5 * Z1, Z1)
print(f'H1:\n{H1}\n')

# Layer 2
Z2 = H1 @ W2.T
print(f'Z2 pre bias:\n{Z2}\n')
Z2 = Z2 + B2
print(f'Z2:\n{Z2}\n')
H2 = np.maximum(0.5 * Z2, Z2)
print(f'H2:\n{H2}\n')
```

## Loss Calculation

```python
# MSE loss calculation (not needed for training)
loss = np.mean((Y - H2) ** 2)
print(f'{loss:.4f}')
```

## Backwards Pass

```python
# Long Chain Calculation
dL_dH2 = 2*(H2 - Y)/len(Y)   # Derivative of loss with respect to H2
print(f'dL/dH2:\n{dL_dH2}\n')

dL_dZ2 = np.multiply(dL_dH2, np.where(Z2 > 0, 1, 0.5))
print(f'dL/dZ2:\n{dL_dZ2}\n')

dL_dH1 = dL_dZ2 @ W2
print(f'dL/dH1:\n{dL_dH1}\n')

dL_dZ1 = np.multiply(dL_dH1, np.where(Z1 > 0, 1, 0.5))
print(f'dL/dZ1:\n{dL_dZ1}\n')

# This one is not needed for training
dL_dX = dL_dZ1 @ W1
print(f'dL/dX:\n{dL_dX}\n')
```

```python
# Leaf Node Calculations
dL_dW2 = dL_dZ2.T @ H1
print(f'dL/dW2:\n{dL_dW2}\n')

dL_dB2 = dL_dZ2.sum(axis=0)
print(f'dL/dB2:\n{dL_dB2}\n')

dL_dW1 = dL_dZ1.T @ X
print(f'dL/dW1:\n{dL_dW1}\n')

dL_dB1 = dL_dZ1.sum(axis=0)
print(f'dL/dB1:\n{dL_dB1}\n')
```

## Update Parameters

```python
W1_new = W1 - lr * dL_dW1
print(f'W1_new:\n{W1_new}\n')

B1_new = B1 - lr * dL_dB1
print(f'B1_new:\n{B1_new}\n')

W2_new = W2 - lr * dL_dW2
print(f'W2_new:\n{W2_new}\n')

B2_new = B2 - lr * dL_dB2
print(f'B2_new:\n{B2_new}\n')
```

```python
# Leaf Node Calculations (Tiling)
dL_dW2 = dL_dZ2[:2].T @ H1[:2]
print(f'dL/dW2:\n{dL_dW2}\n')

# Update W2
W2_new_tiled = W2 - lr * dL_dW2
print(f'W2_new:\n{W2_new_tiled}\n')

# Leaf Node Calculations (Tiling)
dL_dW2 = dL_dZ2[2:4].T @ H1[2:4]
print(f'dL/dW2:\n{dL_dW2}\n')

# Update W2
W2_new_tiled = W2_new_tiled - lr * dL_dW2
print(f'W2_new:\n{W2_new_tiled}\n')


dL_dW1 = dL_dZ1[:2].T @ X[:2]
print(f'dL/dW1:\n{dL_dW1}\n')

# Update W1
W1_new_tiled = W1 - lr * dL_dW1
print(f'W1_new:\n{W1_new_tiled}\n')

# Leaf Node Calculations (Tiling)
dL_dW1 = dL_dZ1[2:4].T @ X[2:4]
print(f'dL/dW1:\n{dL_dW1}\n')

# Update W1
W1_new_tiled = W1_new_tiled - lr * dL_dW1
print(f'W1_new:\n{W1_new_tiled}\n')
```

## Hardware Co-Verification

Validate the **systolic array RTL** (`systolic.sv`) against this numpy golden model.

The 2×2 systolic array computes `Z1 = X @ W1ᵀ` (pre-bias) in **Q8.8 fixed-point**:

| Signal | Meaning |
|---|---|
| `sys_data_out_21` | Column-1 output  →  `Z1[:, 0]` |
| `sys_data_out_22` | Column-2 output  →  `Z1[:, 1]` |

Weight layout in hardware (loaded via `sys_accept_w`):
- **pe11** ← `W1[0][0]`,  **pe21** ← `W1[0][1]`  (column-1, staggered)
- **pe12** ← `W1[1][0]`,  **pe22** ← `W1[1][1]`  (column-2, staggered)


```python
# ── Q8.8 Fixed-Point Helpers (mirrors tiny-tpu fixedpoint.sv) ──────────────
FRAC = 8
SCALE = 1 << FRAC   # 256

def f2q(x):
    """Float → Q8.8 signed 16-bit integer (Python int, sign-extended)."""
    v = int(round(x * SCALE)) & 0xFFFF
    return v - 0x10000 if v >= 0x8000 else v

def q2f(v):
    """Q8.8 signed 16-bit integer → float."""
    v = int(v) & 0xFFFF
    return (v - 0x10000 if v >= 0x8000 else v) / SCALE

def q2hex(v):
    """Q8.8 signed 16-bit integer → 4-digit hex string."""
    return f"0x{int(v) & 0xFFFF:04X}"

def pe_mac(data_in, weight, psum=0.0):
    """
    Single PE MAC in Q8.8:
      psum_out = psum_in + (data_in * weight)
    Matches pe.sv: product = (a_reg * b_reg) >>> FRAC_BITS, then add psum.
    """
    d = f2q(data_in)
    w = f2q(weight)
    p = f2q(psum)
    product = (d * w) >> FRAC          # Q16.16 → Q8.8
    result   = (p + product) & 0xFFFF  # 16-bit truncate
    return q2f(result)                 # back to float for readability

print("Q8.8 helpers loaded.")
print(f"  Example: f2q(0.2985) = {f2q(0.2985)}  (0x{f2q(0.2985)&0xFFFF:04X})")
print(f"  Example: q2f(0x{f2q(0.2985)&0xFFFF:04X}) = {q2f(f2q(0.2985)):.4f}")
```

```python
# ── Systolic Array Golden Model ─────────────────────────────────────────────
# Simulates the 2×2 systolic array computing Z1 = X @ W1ᵀ sample-by-sample.
#
# For each sample i, inputs are staggered by 1 cycle:
#   Cycle 0 → pe11 sees X[i][0],   pe12 sees X[i][0]  (psum_in = 0)
#   Cycle 1 → pe21 sees X[i][1],   pe22 sees X[i][1]  (psum_in from row-1 PE)
#
# Column-1:  sys_data_out_21[i] = pe21 psum = X[i,0]*W1[0,0] + X[i,1]*W1[0,1]
# Column-2:  sys_data_out_22[i] = pe22 psum = X[i,0]*W1[1,0] + X[i,1]*W1[1,1]

Z1_pre_bias = X @ W1.T   # numpy reference

hw_Z1   = []   # Q8.8-simulated floating-point result
hw_int  = []   # raw Q8.8 integers (what the waveform shows)
hw_hex  = []   # hex strings for waveform / simulator comparison

for i in range(len(X)):
    # Column 1: pe11 → pe21
    c1 = pe_mac(X[i][0], W1[0][0])            # pe11
    c1 = pe_mac(X[i][1], W1[0][1], c1)        # pe21 accumulates

    # Column 2: pe12 → pe22
    c2 = pe_mac(X[i][0], W1[1][0])            # pe12
    c2 = pe_mac(X[i][1], W1[1][1], c2)        # pe22 accumulates

    hw_Z1.append([c1, c2])
    hw_int.append([f2q(c1) & 0xFFFF, f2q(c2) & 0xFFFF])
    hw_hex.append([q2hex(f2q(c1)), q2hex(f2q(c2))])

hw_Z1 = np.array(hw_Z1)

print("=" * 60)
print(f"{'Sample':<8} {'numpy Z1[:,0]':>14} {'HW Z1[:,0]':>12}  │  {'numpy Z1[:,1]':>14} {'HW Z1[:,1]':>12}")
print("-" * 60)
for i in range(len(X)):
    print(f"X[{i}]    {Z1_pre_bias[i,0]:>14.4f} {hw_Z1[i,0]:>12.4f}  │  "
          f"{Z1_pre_bias[i,1]:>14.4f} {hw_Z1[i,1]:>12.4f}")

print("\nExpected RTL output (Q8.8 hex — read off waveform / cocotb):")
print(f"{'Sample':<8} {'sys_data_out_21':>17} {'sys_data_out_22':>17}")
print("-" * 45)
for i in range(len(X)):
    print(f"X[{i}]    {hw_hex[i][0]:>17} {hw_hex[i][1]:>17}")
```

```python
# ── Quantization Error Check ────────────────────────────────────────────────
# Q8.8 has a resolution of 1/256 ≈ 0.0039.
# Any difference > 0.0039 means a logic bug, not just rounding.

TOL = 1.5 / SCALE   # 1.5 LSBs tolerance (generous, accounts for rounding both inputs)

errors      = np.abs(Z1_pre_bias - hw_Z1)
max_err     = errors.max()
all_within  = bool((errors <= TOL).all())

print(f"Q8.8 resolution (1 LSB) : {1/SCALE:.6f}")
print(f"Tolerance (1.5 LSBs)    : {TOL:.6f}")
print(f"Max quantization error  : {max_err:.6f}")
print()
if all_within:
    print("✓  All samples within tolerance — golden model matches hardware arithmetic.")
else:
    print("✗  Some samples exceed tolerance — investigate!")
    bad = np.argwhere(errors > TOL)
    for idx in bad:
        i, j = idx
        print(f"   Sample {i}, col {j}: numpy={Z1_pre_bias[i,j]:.4f}  hw={hw_Z1[i,j]:.4f}  "
              f"err={errors[i,j]:.6f}")

print()
print("Error matrix (numpy vs Q8.8 HW):")
print(errors)
```
